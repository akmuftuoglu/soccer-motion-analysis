```python
#Accelerometer calibration

#2sets
z_neg = data[:10,4]
y = data[:10,3] #abritrarly chosen between x and y

def calibration_curve(x1,y1,x2,y2):
  slope = (y2-y1) / (x2-x1)
  intercept = y1 - slope*x1
  return slope, intercept

#Find slope/intercept provided two values
x1 = y[5]
y1 = 0
x2 = z_neg[5]
y2 = -9.81
slope, intercept = calibration_curve(x1,y1,x2,y2)

acceleration = slope*-data[:,4] + intercept
#plt.plot(z_neg, acceleration)
relative_acceleration = acceleration + 15.5
plt.plot(relative_acceleration)
```

```python
#9 trials each

#max ball acc
max_ball_vel = [3.42, 3.58, 4.73,   8.62, 7.62, 6.02,   9.15, 9.02]  # m/s
max_ball_vel_avg = [np.mean([3.42, 3.58, 4.73]), np.mean([8.62, 7.62, 6.02]), np.mean([6.14, 9.15, 9.02])]


#max leg acc
max_leg_acc = [11.48, 10.70, 10.52,  11.55, 12.46, 13.61,   14.49, 15.42]  # m/s^2


#max leg vel
max_leg_vel = [2.95, 5.26, 9.46]
```

```python
import numpy.polynomial.polynomial as poly
coefs = poly.polyfit(max_leg_acc, max_ball_vel, 1)
ffit = poly.polyval(max_leg_acc, coefs)
plt.plot(max_leg_acc, ffit)

max_ball_vel_all = [3.42, 3.58, 4.73,   8.62, 7.62, 6.02,   6.14, 9.15, 9.02]
max_leg_acc_all = [11.48, 10.70, 10.52,  11.55, 12.46, 13.61,  19.09, 14.49, 15.42]
plt.scatter(max_leg_acc_all, max_ball_vel_all)

plt.ylabel('Ball Max Velocity (m/s)')
plt.xlabel('Leg Max Acc (m/s^2)')
plt.title('Leg Acceleration effects on Ball Velocity')

print("Slope: ", coefs[1])
#print("Mean of Leg Acc:", np.mean(max_leg_acc))
#print("Proportion of slope to.. Mean of Leg Acc:", coefs[1] / np.mean(max_leg_acc))

# there is an incease of 0.96 m/s Ball Velocity foe every additional 1 m/s^2 of Leg Max Acc
```

```python
import numpy.polynomial.polynomial as poly
coefs = poly.polyfit(max_leg_vel, max_ball_vel_avg, 1)
ffit = poly.polyval(max_leg_vel, coefs)
plt.plot(max_leg_vel, ffit)

plt.scatter(max_leg_vel, max_ball_vel_avg)

plt.ylabel('Ball Max Velocity (m/s)')
plt.xlabel('Leg Max Velocity (m/s)')
plt.title('Leg Velocity effects on Ball Velocity')

print("Slope: ", coefs[1])
#print("Mean of Leg Vel:", np.mean(max_leg_vel))
#print("Proportion of slope to.. Mean of Leg Vel:", coefs[1] / np.mean(max_leg_vel))

# there is an incease of 0.58 m/s Ball Velocity foe every additional 1 m/s of Leg Max Vel
```
